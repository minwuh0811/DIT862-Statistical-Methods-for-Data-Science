```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import mixture
ns = 100
s = 10
# generate spherical data 

data1 = np.random.randn(ns, 2)
data2 = np.random.randn(ns, 2) + np.array([s, 0])
data3 = np.random.randn(ns, 2) + np.array([0, s])


# concatenate 
data = np.vstack([data1,data2,data3])
# You may plot the data to see how it looks like 
fig=plt.figure(figsize=(19,20))
ax1=fig.add_subplot(1,2,1)
#plot the data
ax1.scatter(data[:, 0], data[:, 1], .8)
K = 5
gmm = mixture.GaussianMixture(n_components=K, n_init=10)
gmm.fit(data)
y_pred=gmm.predict(data)
colors=[]
for i in y_pred:
    if i==0:
        colors.append('#377eb8')
    elif i==1:
        colors.append('#ff7f00')
    elif i==2:
        colors.append('#4daf4a')
    elif i==3:
        colors.append('#f781bf')
    elif i==4:
        colors.append('#a65628')
# plot the clusterd data using different color 
ax2=fig.add_subplot(1,2,2)
ax2.scatter(data[:, 0], data[:, 1], color=colors)
```

# 1. Describe a procedure for estimating the correct number of clusters via AIC and BIC. 

1. calculate the Negative Log-Likelihood of data 𝑁∗𝑔𝑚𝑚.𝑠𝑐𝑜𝑟𝑒(𝑑𝑎𝑡𝑎) where N is the total number of objects (data points).
2. AIC = -ln p(X|.) + ck and BIC = -ln p(X|.) + ck*lnN/2 calculate ck = K*D + (K - 1) + K*D*(D + 1)/2
3. calculate AIC and BIC
4. the smallest number of AIC and BIC is the correct number of clusters

```python
     
def ck(K, D=2):
    return K*D + (K-1) + K*D* (D + 1)/2
def AIC(negLogLikelihood,ck):
    return negLogLikelihood+ck
def BIC(negLogLikelihood,ck):
    return negLogLikelihood+ck/2*np.log(len(data))

aic=[]
bic=[]
totLikelihood=[]
Ncluster=[]
for i in range(1,10):
    gmm = mixture.GaussianMixture(n_components=i, n_init=10)
    gmm.fit(data)
    negLogLikelihood=-len(data)*gmm.score(data)   
    totLikelihood.append(negLogLikelihood)
    aic.append(AIC(negLogLikelihood,ck(i)))
    bic.append(BIC(negLogLikelihood,ck(i)))
    Ncluster.append(i)
plt.plot(Ncluster,totLikelihood,color='olive',marker='o', markerfacecolor='olive', markersize=4, linewidth=2,label='-loglikelihood')
plt.plot(Ncluster,aic,color='skyblue',marker='o', markerfacecolor='skyblue', markersize=4, linewidth=2,label='AIC')
plt.plot(Ncluster,bic,color='purple',marker='o', markerfacecolor='purple', markersize=4, linewidth=2,label='BIC')
plt.legend()

```

# 3. Set 𝑠 = 10 and 𝑛𝑠=500. Assume that you know the covariance matrices are diagonal but not necessarily the same. Then, how would you apply AIC and BIC? Show/plot the results.

1. calculate the Negative Log-Likelihood of data 𝑁∗𝑔𝑚𝑚.𝑠𝑐𝑜𝑟𝑒(𝑑𝑎𝑡𝑎) where N is the total number of objects (data points).
2. AIC = -ln p(X|.) + ck and BIC = -ln p(X|.) + ck*lnN/2 because covariance matrices are known to be diagonal where ck = K*D + (K - 1) 
3. calculate AIC and BIC
4. the smallest number of AIC and BIC is the correct number of clusters

```python
#a s=10 and ns =500
ns = 500
s = 10
# generate spherical data 

data1 = np.random.randn(ns, 2)
data2 = np.random.randn(ns, 2) + np.array([s, 0])
data3 = np.random.randn(ns, 2) + np.array([0, s])
# concatenate 
data = np.vstack([data1,data2,data3])
# You may plot the data to see how it looks like 
fig=plt.figure(figsize=(19,20))
ax1=fig.add_subplot(2,2,1)
#plot the data
ax1.scatter(data[:, 0], data[:, 1], .8)
ax2=fig.add_subplot(2,2,2)
def ckDiag(K, D=2):
    return K*D + (K-1)
def AIC(negLogLikelihood,ck):
    return negLogLikelihood+ck
def BIC(negLogLikelihood,ck):
    return negLogLikelihood+ck/2*np.log(len(data))
aic=[]
bic=[]
totLikelihood=[]
Ncluster=[]
for i in range(1,10):
    gmm = mixture.GaussianMixture(n_components=i,covariance_type='diag', n_init=10)
    gmm.fit(data)
    negLogLikelihood=-len(data)*gmm.score(data)   
    totLikelihood.append(negLogLikelihood)
    aic.append(AIC(negLogLikelihood,ckDiag(i)))
    bic.append(BIC(negLogLikelihood,ckDiag(i)))
    Ncluster.append(i)
ax2.plot(Ncluster,totLikelihood,color='olive',marker='o', markerfacecolor='olive', markersize=4, linewidth=2,label='-loglikelihood')
ax2.plot(Ncluster,aic,color='skyblue',marker='o', markerfacecolor='skyblue', markersize=4, linewidth=2,label='AIC')
ax2.plot(Ncluster,bic,color='purple',marker='o', markerfacecolor='purple', markersize=4, linewidth=2,label='BIC')
ax2.legend()
```

The smallest values of AIC and BIC correspond to the correct number of clusters which is 3. So the correct number of cluster is 3.

# 4. Apply BIC to analyze the number of clusters for a real-world dataset called breast_cancer. The information about the dataset can be found at: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html#sklearn.datasets.load_breast_cancer  

a. Investigate BIC for finding the number of clusters where you assume 
i) you estimate all the covariance matrices separately
ii) you estimate all the full covariance matrices separately, but assume that they are diagonal.

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer().data
bicFull=[]
bicDiag=[]
totLogLikelihoodFull=[]
totLogLikelihoodDiag=[]
Ncluster=[]
def ckDiag(K, D=2):
    return K*D + (K-1)
for i in range(1,50):
    gmmFull = mixture.GaussianMixture(n_components=i,covariance_type='full', n_init=10)
    gmmFull.fit(data)
    negLogLikelihoodFull=-len(data)*gmmFull.score(data)
    totLogLikelihoodFull.append(negLogLikelihoodFull)
    bicFull.append(BIC(negLogLikelihoodFull,ck(i,len(data[0]))))
    gmmDiag = mixture.GaussianMixture(n_components=i,covariance_type='diag', n_init=10)
    gmmDiag.fit(data)
    negLogLikelihoodDiag=-len(data)*gmmDiag.score(data) 
    totLogLikelihoodDiag.append(negLogLikelihoodDiag)
    bicDiag.append(BIC(negLogLikelihoodDiag,ckDiag(i,len(data[0]))))
    Ncluster.append(i)
```

  b. Plot the BIC and the negative log-likelihood results for each case. 

```python
fig=plt.figure(figsize=(19,20))
ax1=fig.add_subplot(2,2,1)
ax1.plot(Ncluster,bicFull,color='olive',marker='o', markerfacecolor='olive', markersize=4, linewidth=2,label='BIC-Full')
ax1.plot(Ncluster,totLogLikelihoodFull, color='skyblue', marker='o', markerfacecolor='skyblue', markersize=4, linewidth=2,label='-loglikelihood-Full'  )
ax1.legend()
ax2=fig.add_subplot(2,2,2)
ax2.plot(Ncluster,bicDiag,color='purple',marker='o', markerfacecolor='purple', markersize=4, linewidth=2,label='BIC-Diag')
ax2.plot(Ncluster,totLogLikelihoodDiag, color='red', marker='o', markerfacecolor='red', markersize=4, linewidth=2,label='-loglikelihood-Diag'  )
ax2.legend()
```
